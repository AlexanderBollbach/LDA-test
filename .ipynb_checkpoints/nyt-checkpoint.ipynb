# ingest Dataset
---

```python
def nyt_docs():
    curr_doc = ""
    file = open('./nyt.txt')
    for line in file:
        if line is '\n' or line.startswith("URL:"):
            if curr_doc != "":
                yield curr_doc
                curr_doc = ""
        else:
            if curr_doc is "":
                curr_doc = line
            else:
                curr_doc += line
                
                
def nyt_docs_firstN(n):
    count = 0
    for doc in nyt_docs():
        count += 1
        yield doc
        if count > n:
            break
```

# Spacy Pipeline

takes a "document" (string) returns tokrns

```python
from spacy.lang.en import English
import spacy

nlp = spacy.load('en')

def token_is_ok(spacy_word):
        return not (spacy_word.is_stop or spacy_word.is_punct or spacy_word.like_num)
            
def spacy_process(docs):
    tokens = []
    for doc in docs:
        spacy_doc = nlp(doc)
        for w in spacy_doc:
            if token_is_ok(w):
                tokens.append(w.lemma_)
        yield tokens
        tokens = []
```

```python
some_docs = list(spacy_process(nyt_docs_firstN(10)))
```

# Dictionary

maintain id->work mapping and create BOW vectors

```python
from gensim.corpora import Dictionary

for doc in some_docs:
    print(doc)

nyt_dict = Dictionary()

nyt_dict.add_documents(some_docs)

        
```

# Phrases

lets look for trigrams with 2nd order phrase application

```python
from gensim import corpora
from gensim.models.phrases import Phrases, Phraser

bigram_model = Phrases(some_docs)
some_docs = [bigram_model[doc] for doc in some_docs]
some_docs = [bigram_model[doc] for doc in some_docs]

for doc in some_docs:
    print(doc)
```

```python
class NYT_Corpus(object):
    def __iter__(self):
        for doc in some_docs:
            print("ddd")
            yield nyt_dict.doc2bow(doc)
```

```python
from gensim.models import LsiModel, LdaModel

model = LdaModel(corpus=NYT_Corpus(), id2word=nyt_dict, num_topics=2)

print(model.show_topics())
```
